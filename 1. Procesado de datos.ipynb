```python
#Librerias
import pandas as pd
import numpy as np
import os
```

```python
#Rutas
path = 'Data/'
footballpath = 'Data/football/ligasprincipales/'
footballotherspath = 'Data/football/otrasligas/'
basketballpath = 'Data/basketball/'
tennismenpath = 'Data/tennis/men/'
tenniswomenpath = 'Data/tennis/women/'
```

```python
#Columnas
footballcolumns = ['Div', 'Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 'IWH', 'IWA']
tennismencolumns = ['ATP', 'Location', 'Tournament', 'Date', 'Series', 'Court', 'Surface','Round', 'Best of', 'Winner', 'Loser', 'B365W', 'B365L']
tenniswomencolumns = ['WTA', 'Location', 'Tournament', 'Date', 'Tier', 'Court', 'Surface','Round', 'Best of', 'Winner', 'Loser', 'B365W', 'B365L']
footballotherscolumns =['Country', 'League', 'Season', 'Date', 'Time', 'Home', 'Away', 'HG','AG', 'Res', 'PH', 'PD', 'PA']
```

```python
#Nuevo Directorio
os.makedirs("Data/Depurado")
```

```python
#Files
footballfiles = os.listdir(path=footballpath)
footballothersfiles = os.listdir(path=footballotherspath)
basketballfiles = os.listdir(path=basketballpath)
tennismenfiles = os.listdir(path=tennismenpath)
tenniswomenfiles = os.listdir(path=tenniswomenpath)
```

```python
#Football
vacio = pd.DataFrame()

for file in footballfiles:
    tempf1 = pd.read_excel(footballpath+file, sheet_name=None)
    sheets = list(tempf1.keys())
    for sheet in sheets:
        tempf2 = pd.read_excel(footballpath+file, sheet_name=sheet)
        tempf2 = tempf2[footballcolumns]
        vacio = vacio.append(tempf2)
        
vacio = vacio.drop_duplicates(keep=False)
vacio = vacio.dropna(how='any')

vacio.to_excel('Data/Depurado/football.xlsx')
```

```python
#Football Others
vacio = pd.DataFrame()

for file in footballothersfiles:
    tempfo1 = pd.read_excel(footballotherspath+file, sheet_name=None)
    sheets = list(tempfo1.keys())
    for sheet in sheets:
        tempfo2 = pd.read_excel(footballotherspath+file, sheet_name=sheet)
        tempfo2 = tempfo2[footballotherscolumns]
        vacio = vacio.append(tempfo2)
        
vacio = vacio.drop_duplicates(keep=False)
vacio = vacio.dropna(how='any')

vacio.to_excel('Data/Depurado/footballothers.xlsx')
```

```python
#Basketball
vacio = pd.DataFrame()

for file in basketballfiles:
    tempb1 = pd.read_excel(basketballpath+file, sheet_name=None)
    sheets = list(tempb1.keys())
    for sheet in sheets:
        tempb2 = pd.read_excel(basketballpath+file, sheet_name=sheet)
        vacio = vacio.append(tempb2)
        
vacio = vacio.drop_duplicates(keep=False)
vacio = vacio.dropna(how='any')
        
vacio.to_excel('Data/Depurado/basketball.xlsx')
```

```python
#TennisMen
vacio = pd.DataFrame()

for file in tennismenfiles:
    temptm1 = pd.read_excel(tennismenpath+file, sheet_name=None)
    sheets = list(temptm1.keys())
    for sheet in sheets:
        temptm2 = pd.read_excel(tennismenpath+file, sheet_name=sheet)
        temptm2 = temptm2[tennismencolumns]
        vacio = vacio.append(temptm2)
        
vacio = vacio.drop_duplicates(keep=False)
vacio = vacio.dropna(how='any')
        
vacio.to_excel('Data/Depurado/tennismen.xlsx')
```

```python
#TennisWomen
vacio = pd.DataFrame()

for file in tenniswomenfiles:
    temptw1 = pd.read_excel(tenniswomenpath+file, sheet_name=None)
    sheets = list(temptw1.keys())
    for sheet in sheets:
        temptw2 = pd.read_excel(tenniswomenpath+file, sheet_name=sheet)
        temptw2 = temptw2[tenniswomencolumns]
        vacio = vacio.append(temptw2)
        
vacio = vacio.drop_duplicates(keep=False)
vacio = vacio.dropna(how='any')
        
vacio.to_excel('Data/Depurado/tenniswomen.xlsx')
```

```python
#Juntar Tennis masculino con Tennis femenino con las columnas necesarias

tennisall = tennismen.append(tenniswomen)
tennisall = tennisall[['Tournament', 'Date','Winner', 'B365W', 'B365L']]
tennisall['Sport'] = 'Tennis'
```

```python
tennisall.head()
```

```python
tennisall.dtypes
```

```python
#Convertir tipo de una columna a float

tennisall['B365L'] = tennisall['B365L'].convert_objects(convert_numeric=True)
```

```python
#Determinar si gana el favorito

tennisall['GanaFavorito'] = tennisall['B365W'] < tennisall['B365L']
tennisall.loc[tennisall['GanaFavorito'] == False, 'GanaFavorito'] = 0
tennisall.loc[tennisall['GanaFavorito'] == True, 'GanaFavorito'] = 1
```

```python
#Cambiar nombre las columnas

tennisall = tennisall.rename(columns={"B365W":"OddW", "B365L":"OddL",'Winner':'Favorito'})
tennisall['CuotaFavorito'] = tennisall[['OddW','OddL']].min(axis=1)
```

```python
tennisall = tennisall[['Tournament','Date','Sport','Favorito','GanaFavorito','CuotaFavorito']]
```
